```python
import ephem
import calendar
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt

from math import radians as rad, degrees as deg
```

```python
g = ephem.Observer()
g.name = 'Los Angeles'
g.lat=rad(34.0211)
g.long=rad(-118.3965)
m = ephem.Moon()

date_list = pd.date_range('2019-06-01', '2019-12-01', freq='D')
date_list = date_list.tolist()

mphase_list = []
for ii in range(0, len(date_list)):
    g.date = date_list[ii]
    g.date -= ephem.hour
    m.compute(g)
    mphase_list.append(m.phase)

df = pd.DataFrame({'moon phase': mphase_list}, index = date_list)
print(df)

df.plot()
plt.show()
```
